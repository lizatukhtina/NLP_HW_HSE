С помощью регулярных выражение очистите текст от тэгов (мета информацию внутри тэгов тоже удалите). В получившемся тексте точки не должны слипаться с началом следующего предложений, также не должно быть лишних пробелов.

```python
import re

def remove_tags(input_text : str) -> str:
    # паттерн для поиска всевозможноных тегов
    find_tags_ptrn = re.compile(r'<[^>]*>')
    # паттерн для поиска "слипаний" двух предложений
    find_merges_ptrn = re.compile(r'([.?!])([А-Я0-9]+)')
    
    # чистка текста от тегов
    clean_text = re.sub(find_tags_ptrn, '', input_text )
    # удалений слипаний
    clean_text = re.sub(find_merges_ptrn, r'\g<1> \g<2>', clean_text)
    # удаление переносов строк 
    clean_text = clean_text.replace('\n', ' ')[1:]
    
    return(clean_text)

text = remove_tags(text)
print(text) # результат
```

Для пайморфи используйте токенизацию из mystem, чтобы исключить влияние токенизации на результат.

### Лемматизация с PyMystem3

```python
from pymystem3 import Mystem
from razdel import tokenize
import json

m = Mystem()
    
# функция для токенизации текста с помощью mystem
def tokenize_via_mystem(text):
    tokens = []
    morph_items = m.analyze(text)
    for item in morph_items:
        tokens.append(item['text'])
        
    return(tokens)

mystem_lemmas = ''.join(m.lemmatize(text))
```

### Лемматизация с pymorphy2

```python
from pymorphy2 import MorphAnalyzer

morph = MorphAnalyzer()
mystem_tokens = tokenize_via_mystem(text)
words_analyzed = [morph.parse(token) for token in mystem_tokens]

pymorphy_lemmas = ''
for w in words_analyzed:
    pymorphy_lemmas += w[0].normal_form
```

### Сравнение разборов со снятием лексической многозначности и без

Сравним разборы с дизамбигуацией и без для слова "**данные**" из следующего контекста:

*Сервис позволяет отслеживать активность определенных пользователей, и выводить эти **данные** в различных формах.*

(строка **63** в таблице ниже)

При поиске леммы для слова "данные" без снятия многозначности, Mystem не учитывал контекст и вывел всевозможные варианты лемм: *данные, давать, данный*. Во втором случае, Mystem учел контекст и выдал только 2 варианта разбора, отбросив варианты, неприменимые для рассматриваемого предложения.

Другой пример — разбор союза "тем" из двойного союза «чем… тем» (см. строка **72** в таблице). Без снятия многозначности в список лемм были включены такие варианты, как местоимение "тот" и существительное "тема". При снятии многозначности, Mystem выдал только один правильный вариант разбора.

**Вывод**: контекстная дизамбигуация в Mystem в разы повышает шансы определить правильный разбор слова 👍 

```python
data = []

# цикл по двум разборам
for s1, s2 in zip(text_parsed[0], text_parsed_d[0]):
        # ключа 'analysis' нет у пробелов знаков пунктуации
        if('analysis' in s1 and 'analysis' in s2):
            # исходная словоформа
            word = s1['text']
            
            lemmas1 = [] # леммы без d
            lemmas2 = [] # леммы с d
            
            # вытаскиваем все леммы из разбора без дизамбигуции
            for lemma1 in s1['analysis']:
                lemmas1.append(f"{lemma1['lex']} : {lemma1['gr']}")
            
            # вытаскиваем все леммы из разбора с дизамбигуацией
            for lemma2 in s2['analysis']:
                lemmas2.append(f"{lemma2['lex']} : {lemma2['gr']}")
                
            data.append([word, ', '.join(lemmas1), ', '.join(lemmas2)]) 


frame = pd.DataFrame(data, columns=["Слово", "Разбор без дизамбигуации", "Разбор с дизамбигуацией"])
frame[63:75]
```
